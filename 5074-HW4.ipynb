## The third project: Predictive Model for Movie Review Classification

The link to the third assignment notebook github repo:
https://raw.githubusercontent.com/HanyiW/Hanyi-Wang/main/5074-HW3

For the Stanford SST Sentiment dataset, I tried 4 models. Model 1 is LSTM Model with Additional Layers and Hyperparameter Tuning. Model 2 is the embedding layer and Conv1d layers. Model 3 is CNN Model with glove embeddings. Model 4 is a bidirectional LSTM model. After discussing with my team, I tried the bidirectional LSTM model and it worked better. In the embedding layer, the max words is set to 10000, features to 10, and max length to 40.


```python
sst = pd.read_csv('sstsentiment.csv', index_col=0)
```

```python
sst.sort_values('accuracy', ascending=False)
```

```python
sst.groupby(['optimizer','num_params']).mean().sort_values('accuracy',ascending=False)
```

```python
sst_new = sst.loc[sst['num_params']!=sst['num_params'].max()]
```

```python
import matplotlib.pyplot as plt
plt.scatter(sst_new['num_params'], sst_new['accuracy'], alpha=0.2)
plt.show()
```
